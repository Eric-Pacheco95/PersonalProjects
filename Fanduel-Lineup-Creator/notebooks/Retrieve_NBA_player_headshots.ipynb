Purpose of this notebook is use the <a href='https://nba-players.herokuapp.com/players'>nba-players api</a> to retrieve player headshots for each NBA player that has played between 2016-2020.

```python
import requests
import joblib
```

```python
player_slugs_names = joblib.load('../joblib_objects/player_slugs_names')
player_names = list(player_slugs_names.keys())
```

```python
images_baseurl_endpoint = 'https://nba-players.herokuapp.com/players'

def get_player_image(player_name):
    
    player_name_split = player_name.split(' ')
    first_name = player_name_split[0].lower()
    last_name = player_name_split[1].lower()
    
    try:
        image_request = requests.get(f'{images_baseurl_endpoint}/{last_name}/{first_name}')
        print(image_request.status_code)
        
        if image_request.status_code == 200:
            
            img = image_request.content
            
            file = open(f'../static/images/player_images/{first_name}_{last_name}.png', 'wb') 
            file.write(img)
            file.close()
                
                
    except:
        
        print(first_name, last_name)
```

```python
for player in player_names:
    get_player_image(player)
```
