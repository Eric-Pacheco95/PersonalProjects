```python
import joblib
import pandas as pd
import datetime as datetime
```

```python
#Load good songs playlist track ids and all liked songs features from csv
liked_track_ids = joblib.load('Joblib_Objects/liked_track_ids')

good_songs_track_ids = joblib.load('Joblib_Objects/good_songs_track_ids')

liked_songs_df = pd.read_csv('liked_songs.csv',index_col=0)
```

```python
#Format dataframe for model creation

#If song is from good songs playlist, this will define a favourite song
liked_songs_df['favourite'] = liked_songs_df['track_id'].map(lambda x: 1 if x in good_songs_track_ids else 0)

#Format genres to only have 1 listed
liked_songs_df['genre'] = liked_songs_df['genre'].map(lambda x: x.strip('][').split(', '))
liked_songs_df['genre'] = liked_songs_df['genre'].map(lambda x: x[0])
liked_songs_df['genre'] = liked_songs_df['genre'].map(lambda x: 'Not Available' if x == '' else x)
```
